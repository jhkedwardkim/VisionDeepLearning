```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d') 

sigma = 1 # 표준편차

x = np.linspace(-5, 5, 100)  # 1차원 가우시안 함수
y = (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x)**2 / (2 * sigma**2))

X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))  # 2차원 가우시안 함수 
Z = (1 / (2 * np.pi * sigma**2)) * np.exp(-((Y)**2 + (X)**2) / (2 * sigma**2))

ax1.plot(x,y)
ax2.plot_surface(X, Y, Z)
plt.show()
```
